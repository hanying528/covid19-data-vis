### US Cases and Deaths by State over Time

```python
df_us_hist = pd.read_csv('us_covid_cases_historical.csv', parse_dates=['submission_date'])
df_us_hist.head()
```

```python
# convert date to yyyy-mm-dd format as str to keep consistency
df_us_hist['date'] = df_us_hist['submission_date'].dt.strftime('%Y-%m-%d')
df_us_hist['date'].head()
```

```python
df_us_hist_sub = df_us_hist[['state', 'date', 'new_case', 'new_death', 'tot_cases', 'new_death']]
```

```python
df_us_hist_sub.rename({'new_case': 'new_cases', 'new_death': 'new_deaths',
                      'tot_death': 'total_deaths', 'tot_cases': 'total_cases'}, axis=1, inplace=True)
```

```python
df_us_hist_sub.shape
```

### US Vaccination Data by State over Time

```python
df_us_vac = pd.read_csv('us_covid_data_0331.csv', parse_dates=['Date'])
df_us_vac.head()
```

```python
# convert date to yyyy-mm-dd format as str to keep consistency
df_us_vac['date'] = df_us_vac['Date'].dt.strftime('%Y-%m-%d')
df_us_vac['date'].head()
```

```python
df_us_vac.Location.unique()
```

```python
df_us_vac_sub = df_us_vac[['date', 'Location', 'Administered_Dose1_Pop_Pct']]
df_us_vac_sub.rename({'Location': 'state', 'Administered_Dose1_Pop_Pct': 'people_vaccinated_per_hundred'},
                     axis=1, inplace=True)
```

```python
df_us_vac_sub.shape
```

```python
# merge two dataframes
df_us_hist_vac = df_us_hist_sub.merge(df_us_vac_sub, how='left', on=['state', 'date'])
```

```python
df_us_hist_vac.shape
```

```python
df_us_hist_vac.sort_values(['state', 'date'])
```

```python
df_us_hist_vac.to_csv('us_vac_hist.csv', index=False)
```

### US Vaccination Data by Age and Race Group
**Only take the most recent snapshot**

```python
df_us_vac_demo = pd.read_csv('us_covid_demographics_data_0331.csv', parse_dates=['Date'])
# convert date to yyyy-mm-dd format as str to keep consistency
df_us_vac_demo['date'] = df_us_vac_demo['Date'].dt.strftime('%Y-%m-%d')
df_us_vac_demo['date'].head()
```

```python
df_us_vac_demo.Demographic_category.unique()
```

```python
race_groups = ['Race_eth_unknown', 'Race_eth_NHOther', 'Race_eth_NHBlack', 'Race_eth_NHNHOPI',
              'Race_eth_Hispanic', 'Race_eth_NHMultiracial', 'Race_eth_NHAsian',
              'Race_eth_NHAIAN', 'Race_eth_NHWhite']
df_us_vac_race = df_us_vac_demo[df_us_vac_demo['Demographic_category'].isin(race_groups)]
df_us_vac_race_latest = df_us_vac_race[df_us_vac_race['date']=='2022-03-31']
```

```python
df_us_vac_race_latest.Administered_Dose1_pct_US.sum()
```

```python
df_us_vac_race_latest_sub = df_us_vac_race_latest[['date', 'Demographic_category', 'Administered_Dose1_pct_US']]
df_us_vac_race_latest_sub.rename({'Demographic_category': 'race_group'}, axis=1, inplace=True)

df_us_vac_race_latest_sub
```

```python
df_us_vac_race_latest_sub.to_csv('us_vac_by_race.csv', index=False)
```

```python
df_us_vac_demo.Demographic_category.unique()
```

```python
age_groups = ['Ages_12-15_yrs', 'Ages_75+_yrs', 'Ages_50-64_yrs',
              'Ages_<5yrs', 'Ages_16-17_yrs', 'Ages_5-11_yrs',
             'Ages_40-49_yrs', 'Ages_65-74_yrs', 'Ages_25-39_yrs', 'Ages_18-24_yrs']
df_us_vac_age = df_us_vac_demo[df_us_vac_demo['Demographic_category'].isin(age_groups)]
df_us_vac_age_latest = df_us_vac_age[df_us_vac_age['date']=='2022-03-31']
```

```python
df_us_vac_age_latest.Administered_Dose1_pct_US.sum()
```

```python
df_us_vac_age_latest = df_us_vac_age_latest[['date', 'Demographic_category', 'Administered_Dose1_pct_US']]
df_us_vac_age_latest.rename({'Demographic_category': 'age_group'}, axis=1, inplace=True)

df_us_vac_age_latest
```

```python
df_us_vac_age_latest.to_csv('us_vac_by_age.csv', index=False)
```
